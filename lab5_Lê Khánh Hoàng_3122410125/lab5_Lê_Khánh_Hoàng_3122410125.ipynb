# **1.1.3: Th·ªëng k√™ m√¥ t·∫£: r∆∞·ª£u ƒë·ªè**

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load d·ªØ li·ªáu r∆∞·ª£u ƒë·ªè
wine_data = pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset1/master/winequality-red.csv")

print("üìå Th√¥ng tin d·ªØ li·ªáu r∆∞·ª£u ƒë·ªè:")
print(wine_data.info())
print(wine_data.describe())

# Th·ªëng k√™ m√¥ t·∫£ c∆° b·∫£n
print("\nMean:", wine_data["quality"].mean())
print("Median:", wine_data["quality"].median())
print("Mode:", wine_data["quality"].mode()[0])
print("Variance:", wine_data["quality"].var())
print("Std:", wine_data["quality"].std())
print("Range:", wine_data["quality"].max() - wine_data["quality"].min())
print("Percentile 75%:", np.percentile(wine_data["quality"], 75))
print("IQR:", stats.iqr(wine_data["quality"]))
```

### 1.1.3 ‚Äî Gi·∫£i th√≠ch k·∫øt qu·∫£ (Th·ªëng k√™ m√¥ t·∫£: r∆∞·ª£u ƒë·ªè)

**M·ª•c ti√™u c·ªßa √¥ n√†y:**
- Hi·ªÉu c√°c ƒë·∫∑c tr∆∞ng t·ªïng qu√°t c·ªßa bi·∫øn `quality` v√† c√°c bi·∫øn ƒëo ch·∫•t l√Ω ‚Äî h√≥a h·ªçc trong dataset r∆∞·ª£u ƒë·ªè.
- Ki·ªÉm tra ph√¢n b·ªë, ƒë·ªô ph√¢n t√°n, v√† nh·∫≠n di·ªán c√°c v·∫•n ƒë·ªÅ nh∆∞ l·ªách, ngo·∫°i l·ªá (outliers).

**Nh·ªØng ƒëi·ªÅu c·∫ßn ƒë·ªçc t·ª´ k·∫øt qu·∫£:**
- `wine_data.describe()` cung c·∫•p n, mean, std, min, 25%, 50% (median), 75%, max cho t·ª´ng bi·∫øn. So s√°nh mean v√† median:
  - N·∫øu mean ‚âà median ‚Üí ph√¢n b·ªë g·∫ßn ƒë·ªëi x·ª©ng.
  - N·∫øu mean > median ‚Üí ph√¢n b·ªë l·ªách ph·∫£i (right-skewed).
  - N·∫øu mean < median ‚Üí ph√¢n b·ªë l·ªách tr√°i (left-skewed).
- `mode` (gi√° tr·ªã th∆∞·ªùng g·∫∑p nh·∫•t) cho bi·∫øt ch·∫•t l∆∞·ª£ng ph·ªï bi·∫øn nh·∫•t (th∆∞·ªùng l√† m·ª©c trung b√¨nh).
- `IQR` (interquartile range = Q3 ‚àí Q1) cho bi·∫øt ƒë·ªô tr·∫£i c·ªßa 50% d·ªØ li·ªáu trung t√¢m; boxplot s·∫Ω l√†m n·ªïi b·∫≠t outliers ngo√†i [Q1 ‚àí 1.5¬∑IQR, Q3 + 1.5¬∑IQR].
- `variance` v√† `std` m√¥ t·∫£ m·ª©c ƒë·ªô ph√¢n t√°n: std l·ªõn ‚Üí nhi·ªÅu bi·∫øn thi√™n.

**H∆∞·ªõng ph√¢n t√≠ch ti·∫øp theo (g·ª£i √Ω h√†nh ƒë·ªông):**
- Ki·ªÉm tra ph√¢n b·ªë cho t·ª´ng bi·∫øn input (acidity, sugar, alcohol...) b·∫±ng `sns.histplot` + `kde` v√† `sns.boxplot`.
- T√≠nh ma tr·∫≠n t∆∞∆°ng quan `wine_data.corr()` ƒë·ªÉ ph√°t hi·ªán c√°c bi·∫øn c√≥ t∆∞∆°ng quan m·∫°nh (tr√°nh ƒëa c·ªông tuy·∫øn khi x√¢y d·ª±ng m√¥ h√¨nh).
- N·∫øu nhi·ªÅu bi·∫øn b·ªã l·ªách (skewed), c√¢n nh·∫Øc log-transform ho·∫∑c Yeo-Johnson ƒë·ªÉ chu·∫©n h√≥a tr∆∞·ªõc khi h·ªçc m√°y.
- N·∫øu m·ª•c ti√™u l√† ph√¢n lo·∫°i ch·∫•t l∆∞·ª£ng: quy·∫øt ƒë·ªãnh chuy·ªÉn `quality` th√†nh nh√£n nh·ªã ph√¢n (low/high) ho·∫∑c multi-class (v√≠ d·ª• 3 nh√≥m) d·ª±a tr√™n ph√¢n b·ªë.

**Ghi ch√∫ khi b√°o c√°o:**

- Tr√¨nh b√†y mean/median/mode v√† k·∫øt lu·∫≠n v·ªÅ skewness, outliers.
- ƒê·ªÅ xu·∫•t ti·ªÅn x·ª≠ l√Ω: impute/transform/capping (x·ª≠ l√Ω ngo·∫°i l·ªá).

### 1.1.4 ‚Äî Gi·∫£i th√≠ch k·∫øt qu·∫£ (Th·ªëng k√™ m√¥ t·∫£: d·ªØ li·ªáu b·ªánh ti·ªÉu ƒë∆∞·ªùng)

**M·ª•c ti√™u c·ªßa √¥ n√†y:**
- X√°c ƒë·ªãnh v·∫•n ƒë·ªÅ ch·∫•t l∆∞·ª£ng d·ªØ li·ªáu (nh∆∞ gi√° tr·ªã b·∫±ng 0, missing), m√¥ t·∫£ s·ª± ph√¢n b·ªë c√°c ch·ªâ s·ªë y t·∫ø (Glucose, BloodPressure, BMI, v.v.) v√† ƒë√°nh gi√° bi√™n ƒë·ªô/ƒë·ªô bi·∫øn thi√™n.

**Nh·ªØng v·∫•n ƒë·ªÅ th∆∞·ªùng g·∫∑p v√† c√°ch hi·ªÉu:**
- `Trong dataset Pima/diabetes, nhi·ªÅu c·ªôt y t·∫ø c√≥ gi√° tr·ªã 0 (v√≠ d·ª• Glucose, BloodPressure, SkinThickness, Insulin, BMI). Gi√° tr·ªã 0 th∆∞·ªùng kh√¥ng h·ª£p l·ªá trong ng·ªØ c·∫£nh y t·∫ø v√† bi·ªÉu hi·ªán cho d·ªØ li·ªáu thi·∫øu.
- `describe()` s·∫Ω cho th·∫•y min = 0 cho nh·ªØng c·ªôt n√†y; c·∫ßn ph·∫£i ki·ªÉm tra v·ªõi `(diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] == 0).sum()` ƒë·ªÉ bi·∫øt s·ªë l∆∞·ª£ng gi√° tr·ªã 0.
- ƒê·ªô l·ªách chu·∫©n l·ªõn ·ªü c√°c bi·∫øn nh∆∞ Glucose ho·∫∑c Insulin cho th·∫•y bi·∫øn thi√™n d·ªØ li·ªáu cao ‚Äî l∆∞u √Ω ·∫£nh h∆∞·ªüng t·ªõi vi·ªác scaling v√† l·ª±a ch·ªçn ph∆∞∆°ng ph√°p imputation.

**ƒê·ªÅ xu·∫•t x·ª≠ l√Ω (chi ti·∫øt, c√≥ l√Ω do):**
- Thay c√°c gi√° tr·ªã 0 b·∫±ng NaN cho c√°c c·ªôt y t·∫ø kh√¥ng th·ªÉ c√≥ 0:
`diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0, np.nan)`
- Imputation:
  - N·∫øu t·ªâ l·ªá missing nh·ªè ‚Üí d√πng `median` imputation (kh√¥ng nh·∫°y outliers).
  - N·∫øu t·ªâ l·ªá missing v·ª´a ph·∫£i ‚Üí d√πng KNNImputer ho·∫∑c IterativeImputer ƒë·ªÉ d·ª± ƒëo√°n gi√° tr·ªã d·ª±a tr√™n c√°c bi·∫øn c√≤n l·∫°i.
  - Tr√°nh mean imputation n·∫øu ph√¢n b·ªë l·ªách m·∫°nh (mean b·ªã ·∫£nh h∆∞·ªüng b·ªüi outliers).
- Ki·ªÉm tra ph√¢n b·ªë sau imputation; n·∫øu d·ªØ li·ªáu v·∫´n l·ªách ‚Üí c√¢n nh·∫Øc transform (log/box-cox/yeo-johnson).
- Chu·∫©n h√≥a/scale: nhi·ªÅu thu·∫≠t to√°n c·∫ßn chu·∫©n h√≥a (StandardScaler) ho·∫∑c robust scaling n·∫øu c√≤n outliers.

**G·ª£i √Ω ph√¢n t√≠ch th√™m:**
- Ki·ªÉm tra ph√¢n b·ªë theo nh√£n Outcome (0/1): `diabetes_data.groupby('Outcome').mean()` ƒë·ªÉ th·∫•y kh√°c bi·ªát trung b√¨nh gi·ªØa nh√≥m b·ªánh nh√¢n c√≥/kh√¥ng m·∫Øc.
- Ki·ªÉm tra imbalance c·ªßa Outcome: `diabetes_data['Outcome'].value_counts()`; n·∫øu l·ªách, c√¢n nh·∫Øc SMOTE/undersampling khi x√¢y m√¥ h√¨nh.

### 1.3.3 ‚Äî Gi·∫£i th√≠ch (SweetViz: Marketing Campaign)

**M·ª•c ti√™u:**
- D√πng SweetViz ƒë·ªÉ sinh b√°o c√°o EDA t·ª± ƒë·ªông, nhanh ch√≥ng c√≥ c√°i nh√¨n t·ªïng quan v·ªÅ bi·∫øn, ph√¢n b·ªë, missing, v√† t∆∞∆°ng quan.

**Ph·∫ßn c·∫ßn ki·ªÉm tra trong report SweetViz:**
- Overview:
  - S·ªë h√†ng/s·ªë c·ªôt, s·ªë null, duplicate rows.
- Variables:
  - V·ªõi m·ªói bi·∫øn: distribution (continuous) ho·∫∑c top categories (categorical), skewness, percentiles, distinct count.
- Missing values:
  - X√°c ƒë·ªãnh bi·∫øn c√≥ t·ªâ l·ªá missing l·ªõn (ng∆∞·ª°ng 5‚Äì10% tr·ªü l√™n c·∫ßn ch√∫ √Ω).
- Associations / Pairwise interactions:
  - Bi·ªÉu di·ªÖn m·ªëi quan h·ªá gi·ªØa bi·∫øn v·ªõi bi·∫øn, v√† v·ªõi target (n·∫øu ch·ªâ ƒë·ªãnh).
- Target analysis:
  - N·∫øu b·∫°n ch·ªâ ƒë·ªãnh m·ªôt c·ªôt target (v√≠ d·ª• Income ph√¢n nh√≥m), SweetViz ph√¢n t√≠ch m·ªëi li√™n h·ªá gi·ªØa features v√† target.

**C√°ch s·ª≠ d·ª•ng k·∫øt qu·∫£ ƒë·ªÉ ra quy·∫øt ƒë·ªãnh ti·ªÅn x·ª≠ l√Ω:**
