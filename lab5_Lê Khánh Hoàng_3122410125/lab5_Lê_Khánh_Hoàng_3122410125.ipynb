# **1.1.3: Thống kê mô tả: rượu đỏ**

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load dữ liệu rượu đỏ
wine_data = pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset1/master/winequality-red.csv")

print("📌 Thông tin dữ liệu rượu đỏ:")
print(wine_data.info())
print(wine_data.describe())

# Thống kê mô tả cơ bản
print("\nMean:", wine_data["quality"].mean())
print("Median:", wine_data["quality"].median())
print("Mode:", wine_data["quality"].mode()[0])
print("Variance:", wine_data["quality"].var())
print("Std:", wine_data["quality"].std())
print("Range:", wine_data["quality"].max() - wine_data["quality"].min())
print("Percentile 75%:", np.percentile(wine_data["quality"], 75))
print("IQR:", stats.iqr(wine_data["quality"]))
```

### 1.1.3 — Giải thích kết quả (Thống kê mô tả: rượu đỏ)

**Mục tiêu của ô này:**
- Hiểu các đặc trưng tổng quát của biến `quality` và các biến đo chất lý — hóa học trong dataset rượu đỏ.
- Kiểm tra phân bố, độ phân tán, và nhận diện các vấn đề như lệch, ngoại lệ (outliers).

**Những điều cần đọc từ kết quả:**
- `wine_data.describe()` cung cấp n, mean, std, min, 25%, 50% (median), 75%, max cho từng biến. So sánh mean và median:
  - Nếu mean ≈ median → phân bố gần đối xứng.
  - Nếu mean > median → phân bố lệch phải (right-skewed).
  - Nếu mean < median → phân bố lệch trái (left-skewed).
- `mode` (giá trị thường gặp nhất) cho biết chất lượng phổ biến nhất (thường là mức trung bình).
- `IQR` (interquartile range = Q3 − Q1) cho biết độ trải của 50% dữ liệu trung tâm; boxplot sẽ làm nổi bật outliers ngoài [Q1 − 1.5·IQR, Q3 + 1.5·IQR].
- `variance` và `std` mô tả mức độ phân tán: std lớn → nhiều biến thiên.

**Hướng phân tích tiếp theo (gợi ý hành động):**
- Kiểm tra phân bố cho từng biến input (acidity, sugar, alcohol...) bằng `sns.histplot` + `kde` và `sns.boxplot`.
- Tính ma trận tương quan `wine_data.corr()` để phát hiện các biến có tương quan mạnh (tránh đa cộng tuyến khi xây dựng mô hình).
- Nếu nhiều biến bị lệch (skewed), cân nhắc log-transform hoặc Yeo-Johnson để chuẩn hóa trước khi học máy.
- Nếu mục tiêu là phân loại chất lượng: quyết định chuyển `quality` thành nhãn nhị phân (low/high) hoặc multi-class (ví dụ 3 nhóm) dựa trên phân bố.

**Ghi chú khi báo cáo:**
- Nêu rõ số lượng quan sát, các biến có missing (nếu có).
- Trình bày mean/median/mode và kết luận về skewness, outliers.
- Đề xuất tiền xử lý: impute/transform/capping (xử lý ngoại lệ).

**Câu lệnh hữu ích để kiểm tra thêm:**
- Kiểm tra tần suất:  
  `wine_data['quality'].value_counts().sort_index()`
- Ma trận tương quan:  
  `wine_data.corr()`
- Kiểm tra outliers từng cột:  
  `sns.boxplot(x=wine_data['alcohol'])`


# **1.1.4: Thống kê mô tả: dữ liệu bệnh tiểu đường**

```python
import pandas as pd

# Load dữ liệu tiểu đường
diabetes_data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv")

print("📌 Thông tin dữ liệu tiểu đường:")
print(diabetes_data.info())
print(diabetes_data.describe())
```

### 1.1.4 — Giải thích kết quả (Thống kê mô tả: dữ liệu bệnh tiểu đường)

**Mục tiêu của ô này:**
- Xác định vấn đề chất lượng dữ liệu (như giá trị bằng 0, missing), mô tả sự phân bố các chỉ số y tế (Glucose, BloodPressure, BMI, v.v.) và đánh giá biên độ/độ biến thiên.

**Những vấn đề thường gặp và cách hiểu:**
- `Trong dataset Pima/diabetes, nhiều cột y tế có giá trị 0 (ví dụ Glucose, BloodPressure, SkinThickness, Insulin, BMI). Giá trị 0 thường không hợp lệ trong ngữ cảnh y tế và biểu hiện cho dữ liệu thiếu.
- `describe()` sẽ cho thấy min = 0 cho những cột này; cần phải kiểm tra với `(diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] == 0).sum()` để biết số lượng giá trị 0.
- Độ lệch chuẩn lớn ở các biến như Glucose hoặc Insulin cho thấy biến thiên dữ liệu cao — lưu ý ảnh hưởng tới việc scaling và lựa chọn phương pháp imputation.

**Đề xuất xử lý (chi tiết, có lý do):**
- Thay các giá trị 0 bằng NaN cho các cột y tế không thể có 0:
`diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0, np.nan)`
- Imputation:
  - Nếu tỉ lệ missing nhỏ → dùng `median` imputation (không nhạy outliers).
  - Nếu tỉ lệ missing vừa phải → dùng KNNImputer hoặc IterativeImputer để dự đoán giá trị dựa trên các biến còn lại.
  - Tránh mean imputation nếu phân bố lệch mạnh (mean bị ảnh hưởng bởi outliers).
- Kiểm tra phân bố sau imputation; nếu dữ liệu vẫn lệch → cân nhắc transform (log/box-cox/yeo-johnson).
- Chuẩn hóa/scale: nhiều thuật toán cần chuẩn hóa (StandardScaler) hoặc robust scaling nếu còn outliers.

**Gợi ý phân tích thêm:**
- Kiểm tra phân bố theo nhãn Outcome (0/1): `diabetes_data.groupby('Outcome').mean()` để thấy khác biệt trung bình giữa nhóm bệnh nhân có/không mắc.
- Kiểm tra imbalance của Outcome: `diabetes_data['Outcome'].value_counts()`; nếu lệch, cân nhắc SMOTE/undersampling khi xây mô hình.


# **1.2.2: Trực quan hóa rượu đỏ**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram chất lượng rượu
plt.figure(figsize=(10,5))
sns.histplot(wine_data["quality"], bins=7, kde=True)
plt.title("Phân bố chất lượng rượu đỏ")
plt.show()

# Boxplot chất lượng rượu
plt.figure(figsize=(10,5))
sns.boxplot(x=wine_data["quality"])
plt.title("Boxplot chất lượng rượu đỏ")
plt.show()
```

### 1.2.2 — Giải thích (Trực quan hóa rượu đỏ)

**Mục tiêu của ô này:**
- Dùng biểu đồ để minh hoạ trực quan phân bố quality và phát hiện mẫu/ngoại lệ không dễ thấy từ thống kê mô tả.

**Cách đọc biểu đồ và kết luận:**
- Histogram (sns.histplot): cho biết mật độ và tần suất các giá trị quality. Nếu histogram có một đỉnh (unimodal) quanh 5–6 → majority quality ở mức trung bình. Presence of long tail → skewness.
- KDE line tiện để ước lượng phân phối liên tục; khi kết hợp với histogram giúp thấy rõ shape.
- Boxplot: xác định outliers (điểm nằm ngoài whiskers). Nếu có nhiều outliers, xem xét nguyên nhân (lỗi nhập liệu hay tình trạng thực tế).
- Kết hợp violinplot để thấy mật độ ở từng mức chất lượng nếu muốn so sánh nhiều nhóm.

**Hành động tiếp theo dựa trên đồ thị:**
- Nếu phân bố chất lượng hẹp (chủ yếu 5–6) → cân nhắc chuyển quality thành nhãn coarser (ví dụ low/medium/high) để dễ phân loại.
- Dùng scatter/pairplot giữa quality và các biến vật lý-hoá học (alcohol, acidity, sulfates...) để tìm biến ảnh hưởng mạnh.
- Sử dụng heatmap cho ma trận tương quan để phát hiện biến độc lập/đa cộng tuyến, giúp chọn feature giảm dimension nếu cần.

**Gợi ý báo cáo:**
- Chèn histogram và boxplot với chú giải (axis labels, title). Ghi nhận các phát hiện: vị trí trung tâm, skewness, tồn tại outliers, và đề xuất tiền xử lý.

### 1.3.3 — Giải thích (SweetViz: Marketing Campaign)

**Mục tiêu:**
- Dùng SweetViz để sinh báo cáo EDA tự động, nhanh chóng có cái nhìn tổng quan về biến, phân bố, missing, và tương quan.

**Phần cần kiểm tra trong report SweetViz:**
- Overview:
  - Số hàng/số cột, số null, duplicate rows.
- Variables:
  - Với mỗi biến: distribution (continuous) hoặc top categories (categorical), skewness, percentiles, distinct count.
- Missing values:
  - Xác định biến có tỉ lệ missing lớn (ngưỡng 5–10% trở lên cần chú ý).
- Associations / Pairwise interactions:
  - Biểu diễn mối quan hệ giữa biến với biến, và với target (nếu chỉ định).
- Target analysis:
  - Nếu bạn chỉ định một cột target (ví dụ Income phân nhóm), SweetViz phân tích mối liên hệ giữa features và target.

**Cách sử dụng kết quả để ra quyết định tiền xử lý:**
- Nếu biến numeric có skew > 1 → áp dụng log/box-cox/yeo-johnson trước modeling.
- Nếu categorical có quá nhiều category (high-cardinality) → cân nhắc regroup (top-k + 'Other') hoặc dùng target/frequency encoding.
- Nếu có nhiều missing ở một biến then drop biến đó nếu không quan trọng; nếu quan trọng thì impute phù hợp
- Nếu 2 biến tương quan cao (> 0.8) → giữ một biến hoặc dùng PCA.

**Lưu ý khi dùng SweetViz:**
- SweetViz tạo file HTML lớn; tải về để xem offline.
- Báo cáo rất tốt để trình bày nhanh, nhưng luôn kiểm tra kỹ các phân tích với code thủ công nếu cần số liệu chính xác (ví dụ khi vẽ biểu đồ cho báo cáo chính thức).
